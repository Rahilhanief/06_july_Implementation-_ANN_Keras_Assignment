```python
## Q No. 2 :
import pandas as pd
df=pd.read_csv('wine.csv')
df.head()
```

```python
## Q No.3 :
df.isnull().sum()   # null values
```

```python
df.info()
```

```python
df.describe()
```

```python
#Encoding categorical variable
df['quality_cat'] = df['quality'].astype('category').cat.codes
df.head()
```

```python
#Dropping quality column
df = df.drop('quality',axis=1)
```

```python
df.head()
```

```python
## Q No. 4 :
X=df.iloc[:,:-1] #independent variables
y=df.iloc[:,-1]  # dependent variable
```

```python
## Q No. 5 :
#Creating training and test set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.1,random_state=45)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
## Q No. 6 :
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
## Q No. 7 :
## Q No. 8 :
# Import `Sequential` from `keras.models`
from keras.models import Sequential
 
# Import `Dense` from `keras.layers`
from keras.layers import Dense
 
# Initialize the constructor
model = Sequential()
 
# Add an input layer
model.add(Dense(60, activation ='relu', input_shape =(11,)))
 
# Add one hidden layer
model.add(Dense(60, activation ='relu'))

# Add second hidden layer
model.add(Dense(60, activation ='relu'))

# Add an output layer
model.add(Dense(1, activation ='sigmoid'))
```

```python
## Q No. 9 :
# Model Summary
model.summary()
```

```python
## Q No. 10 :
## Q NO. 11 :
model.compile(loss ='binary_crossentropy',optimizer ='adam', metrics =['accuracy'])
```

```python
## Q NO. 12 :
# Training Model
history=model.fit(X_train,y_train, epochs = 20,batch_size = 2, verbose = 1)
```

```python
## Q No. 13 :
model.get_weights()
```

```python
model.get_compile_config()
```

```python
## Q No. 14 :
import pickle
# save the model as a pickle file
model_pkl_file = "wine_quality_model.pkl"  

with open(model_pkl_file, 'wb') as file:  
    pickle.dump(model, file)
```

```python
## Q No. 15:
import matplotlib.pyplot as plt
print(history.history.keys())
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
## Q No. 16 :
# evaluate the model
scores = model.evaluate(X_test, y_test, verbose=0)
```
